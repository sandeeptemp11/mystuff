```python
import math
limit = 1000000
harmonic_sum = 0
x_values = []
harmonic_sum_values = []
logn_values = []
for n in range(1, limit + 1):
    harmonic_sum += (1/n)
    x_values.append(n)
    harmonic_sum_values.append(harmonic_sum)
    logn_values.append(math.log(n) + 0.5) 
    # if you put 0.57 instead of 0.5, you cannot see two lines!! in plot
    # Thanks to  Eulerâ€“Mascheroni constant !!
```

```python
import matplotlib.pyplot as plt

plt.plot(x_values, harmonic_sum_values, label = "Harmonic sum values")
plt.plot(x_values, logn_values, label = "Log(n) values")
plt.legend()
plt.title("Harmonic number")
plt.xlabel("n->")
plt.show()
```
